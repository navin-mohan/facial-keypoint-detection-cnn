```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# from collections import defaultdict
%matplotlib inline
```

```python
def display_image(row):
    img = np.array([int(x) for x in row['Image'].split()],dtype=int).reshape((96,96))
    plt.imshow(img,cmap='gray')
    cols = row.keys()[:-1]
    coords = [ (row[i],row[i+1]) for i in range(0,len(row)-1,2)]
    x,y = zip(*coords)
    plt.scatter(x,y)
    plt.show()

training_set = pd.read_csv('./dataset/train.csv')

display_image(training_set.iloc[9])
```
